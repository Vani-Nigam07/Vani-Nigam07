6. Make a countplot of Movies directed by top 5 directors.

```python
Top_5_Diros=movies.set_index(['Director']).Gross.sort_values(ascending=False).index[0:5]
top_5_movies=movies.set_index(['Series_Title']).Gross.sort_values(ascending=False).index[0:5]
Serie= pd.Series(top_5_movies, index=[Top_5_Diros])
sns.countplot(x=Top_5_Diros, data=Serie)
```

7. Make a plot to analyze distribution of IMDB rating?

```python

sns.countplot(x='IMDB_Rating', data=movies)
```

8. What type of Genre has the highest IMDB rating?

```python
m=movies.sort_values(['IMDB_Rating'],ascending=False)[0:10]
plt.figure(figsize=(6,6))
g=sns.histplot(y='IMDB_Rating', x='Genre', data=m)
sns.set_context("paper")

plt.show()
```

9. Name the top 5 movies that have earned highest Meta score. 

```python
q=movies.sort_values(['Meta_score'],ascending=False)[0:50]

plt.figure(figsize=(6,6))
g=sns.scatterplot(y='Series_Title', x='Meta_score', data=q)
sns.set_context("paper")

plt.show()
```

10. Of which Genre movie has the highest Runtime?

```python
q=movies.sort_values(['Runtime'],ascending=False)[0:40]

plt.figure(figsize=(6,6))
g=sns.scatterplot(y='Genre', x='Runtime', data=q)
sns.set_context("paper")

plt.show()
```
